```python
def _save_object(obj, file_name):
    """
    Save any python object to file.
    
    :param: obj : - python objest to be saved
    :param: file_name : - output file, wil be overwrite if exists
    """
    with open(file_name,'w') as f:
        pickle.dump(obj,f)

def _load_object(file_name):
    """
    Save any python object to file.
    
    :param: file_name : - output file, wil be overwrite if exists
    :return: obj : - loaded pthon object
    """
    res = None
    with open(file_name,'r') as f:
        res = pickle.load(f)
        
    return res

def mkdir_p(path):
    try:
        os.makedirs(path)
    except OSError as exc: # Python >2.5
        if exc.errno == errno.EEXIST and os.path.isdir(path):
            pass
        else: raise
            
def create_CRL(directory=None, voids_params=None,*args, **keywrds):
    """
    This function build CLR or load it from file.
    Out/input filename builded as sequence of function parameters.
    
    :param: directory: - output directory.
    """
    if not isinstance(voids_params,tuple):
        raise TypeError('Voids_params must be tule')
        
    file_name = '_'.join([str(a) for a in args[:-1]])
    subdir_name = '_'.join([str(v) for v in voids_params])
    if directory is None:
        full_path = os.path.join(subdir_name,file_name+'.pkl')
    else:
        full_path = os.path.join(directory, subdir_name, file_name+'.pkl')
    
    if  os.path.isfile(full_path):
        print 'Found file {}. CLR will be loaded from file'.format(full_path)
        res = _load_object(full_path)
        return res
    else:
        print 'CLR file NOT found. CLR will be recalculated and saved in file {}'.format(full_path)
        res = srwlib.srwl_opt_setup_CRL(*args)
        mkdir_p(os.path.dirname(full_path))
        _save_object(res, full_path)
        return res 
    
```

```python
def calculate_theta_fwhm_cdr(ekev,qnC):
    theta_fwhm = (17.2 - 6.4 * sqrt(qnC))*1e-6/ekev**0.85
    return theta_fwhm
```

```python
def build_voids(nVoidInRectPar, rMinVoid, rMaxVoid ,diamCRL, strDataFolderName):

    baseNx = 201 #(auxiliary) numbers of points in tabulated functions to be used (in rejection method)
    baseNy = 201

    baseYvsXmin = array.array('d', [0]*baseNx)
    baseYvsXmax = array.array('d', [0]*baseNx)
    rCRL = 0.5*diamCRL
    rCRLe2 = rCRL*rCRL
    dx = diamCRL/(baseNx - 1)
    x = -rCRL
    for ix in range(baseNx):
        baseYvsXmin[ix] = -sqrt(abs(rCRLe2 - x*x))
        baseYvsXmax[ix] = -baseYvsXmin[ix]
        x += dx
    baseNpSurf = baseNx*baseNy
    baseZvsXYmin = array.array('d', [0]*baseNpSurf)
    baseZvsXYmax = array.array('d', [0]*baseNpSurf)
    a = 0.5/rMinCRL
    halfWallThickCRL = 0.5*wallThickCRL
    y = -rCRL
    it = 0
    for iy in range(baseNy):
        ye2 = y*y
        x = -rCRL
        for ix in range(baseNx):
            baseZvsXYmin[it] = -halfWallThickCRL - a*(x*x + ye2)
            baseZvsXYmax[it] = -baseZvsXYmin[it]
            it += 1
            x += dx
        y += dx
    arVoidCenCoordInCRL = srwlib.srwl_uti_rand_fill_vol(nVoidInRectPar,-rCRL,rCRL,baseNx,baseYvsXmin,baseYvsXmax,-rCRL,rCRL,baseNy,baseZvsXYmin,baseZvsXYmax)
    numpy.savetxt(''.join([strDataFolderName, '/coord_np',str(nVoidInRectPar),'_ran1_1d.dat']), arVoidCenCoordInCRL,fmt='%1.4e')  
    nc = len(arVoidCenCoordInCRL)
    coordTmp=numpy.array(arVoidCenCoordInCRL).reshape(nc/3,3)
    numpy.savetxt(''.join([strDataFolderName, '/coord_np',str(nVoidInRectPar),'_ran1_xyz.dat']), coordTmp,fmt='%1.4e')

    ##Replacing longitudinal coordinate by random Radii:
    random.seed()
    for i in range(int(len(arVoidCenCoordInCRL)/3)):
        arVoidCenCoordInCRL[3*i + 2] = rMinVoid + (rMaxVoid - rMinVoid)*random.random()

    nc = len(arVoidCenCoordInCRL)
    coordTmp=numpy.array(arVoidCenCoordInCRL).reshape(nc/3,3)
    numpy.savetxt(''.join([strDataFolderName, '/coord_np',str(nVoidInRectPar),'_ran1_xyRvoid.dat']), coordTmp,fmt='%1.4e')
    return arVoidCenCoordInCRL
```

```python
def plot_wfront(mwf,comment):    
    ii=mwf.get_intensity(slice_number=0,polarization='vertical')
    imax = numpy.max(ii)
    #ph=mwf.get_phase(slice=0,polarization='vertical')
    pylab.figure()
    mesh = mwf.params.Mesh
    [xmin, xmax, ymin, ymax]= mesh.xMin, mesh.xMax, mesh.yMin, mesh.yMax 
    pylab.imshow(ii,extent=[xmin*1e3, xmax*1e3, ymin*1e3, ymax*1e3])
    pylab.axis('auto')
    pylab.colorbar()
    pylab.set_cmap('bone')
    pylab.xlabel('x (mm)')
    pylab.ylabel('y (mm)')
    pylab.title('Intensity distribution, '+ comment)   
    
    xa = 1e3*linspace(xmax,xmin,mwf.params.Mesh.nx)
    ya = 1e3*linspace(ymax,ymin,mwf.params.Mesh.ny)
    
    pylab.figure()
    i_y = ii[:,mwf.params.Mesh.nx /2]
    pylab.plot(ya,i_y)
    pylab.xlim(min(ya[numpy.where(i_y > imax/300)]),max(ya[numpy.where(i_y > imax/300)]))
    pylab.xlabel('y (mm)')
    pylab.title('Vertical cut, x=0')   
    pylab.grid(True)
    
    pylab.figure()
    i_x = ii[mwf.params.Mesh.ny /2,:]
    pylab.plot(xa, i_x)
    pylab.xlim(min(xa[numpy.where(i_x > imax/300)]),max(xa[numpy.where(i_x > imax/300)]))
    pylab.xlabel('x (mm)')
    pylab.title('Horizontal cut, y=0') 
    pylab.grid(True)
    
    #pylab.figure()
    #pylab.plot(ya,ph[:,mwf.Radiation.nx /2])
    #pylab.ylim(-numpy.pi, numpy.pi)
    #pylab.xlabel('y (m)')
    #pylab.title('phase, vertical cut, x=0')   
    #pylab.grid(True)
    
    #pylab.figure()
    #pylab.plot(xa,ph[mwf.Radiation.ny /2,:])
    #pylab.xlim(min(xa[numpy.where(i_x > imax/300)]),max(xa[numpy.where(i_x > imax/300)]))
    #pylab.ylim(-numpy.pi, numpy.pi)
    #pylab.xlabel('x (m)')
    #pylab.title('phase, horizontal cut, y=0') 
    #pylab.grid(True)   
    pylab.show()
```

```python
def propagate(src_h5,dst_h5,optBL):
    #print_beamline(optBL)   # 
    print ('*****reading wavefront from h5 file...')
    w1=Wavefront()
    w1.load_hdf5(src_h5)
    wfr1=w1._srwl_wf
    print('propagating wavefront...')   
    srwlib.srwl.PropagElecField(wfr1, optBL)
    print('done')
    w2 = Wavefront(wfr1)
    print('save hdf5: '+dst_h5)
    w2.store_hdf5(dst_h5)
    del w1
    del wfr1
    print('done')
    return w2
```

```python
#**********************Input Parameters and structures:
    
strDataFolderName = 'data_example_07_caseA_0307_iii' #data sub-folder name
if not os.path.exists(strDataFolderName):
            os.mkdir(strDataFolderName)
#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@#
#files_prefix='o9999'

#Longitudinal Position [m] at which Electric Field has to be calculated, 
#i.e. the position of the first optical element
d2waist = 220
ekev = 6.742
qnC = 0.025
theta_fwhm = calculate_theta_fwhm_cdr(ekev,qnC)
k = 2*sqrt(2*log(2))
theta_rms = theta_fwhm / k
print ('theta_fwhms:', theta_fwhm)
sigX = 12.39e-10/(ekev*4*pi*theta_rms) 
print ('waist (sigma), waist_fwhm [m]:',sigX,sigX*k)
firstHorAp = theta_rms*d2waist*6 #First Aperture [m]
firstVertAp = firstHorAp #[m] 

nx = ny = 100

wfr = build_gauss_wavefront_xy(nx,ny,ekev,
                             -0.5*firstHorAp,0.5*firstHorAp,
                             -0.5*firstVertAp, 0.5*firstVertAp,
                             sigX, sigX, d2waist)
```

```python
mwf = Wavefront(wfr)
ip = numpy.floor(ekev)
frac = numpy.floor((ekev - ip)*1e3)
fname0 = 'g' + str(int(ip))+'_'+str(int(frac))+'kev'
fullname0 = strDataFolderName+'/'+fname0+'.h5'
print('save hdf5: '+fullname0)
mwf.store_hdf5(fullname0)
print('done')
plot_wfront(mwf,'at initial position '+str(d2waist)+' m')
fwhm_x = calculate_fwhm_x(mwf)
print ('FWHMx [mm], theta_fwhm [urad], lambda [A]:',fwhm_x*1e3,fwhm_x/d2waist*1e6, 12.39/ekev)
```
